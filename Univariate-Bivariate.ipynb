```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
df.head()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

Univariate

```python
sns.countplot(x="Survived",data=df)
plt.show()
```

```python
sns.countplot(x="Sex",data=df)
plt.show()
```

```python
sns.countplot(x="Embarked",data=df)
plt.show()
```

```python
sns.boxplot(x="Age",data=df)
plt.show()
```

```python
plt.hist("Age",data=df)
plt.show()
```

```python
plt.hist("Pclass",data=df)
plt.show()
```

```python
plt.hist("Parch",data=df)
plt.show()
```

Bivariate

```python
sns.countplot(x="Survived",hue="Sex",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Pclass",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Embarked",data=df)
plt.show()
```

```python
sns.relplot(x="Age",y="Fare",data=df)
```

```python
sns.relplot(x="Parch",y="Fare",data=df)
```

```python
sns.relplot(x="SibSp",y="Fare",data=df)
```

```python
sns.relplot(x="Age",y="Parch",data=df)
```

```python
sns.relplot(x="Age",y="SibSp",data=df)
```

```python
sns.relplot(x="Parch",y="Pclass",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Survived",data=df,kind="violin")
```
